```python
import numpy as np
a = np.array([-3,1])
b = np.array([-1,2])
c = np.array([-1.15300,1])
acdotp = np.dot(a,c)
bcdotp = np.dot(b,c)
mg_a = (a[0]**2 + a[1]**2)**0.5
mg_b = (b[0]**2 + b[1]**2)**0.5
mg_c = (c[0]**2 + c[1]**2)**0.5
cosx = acdotp/(mg_a*mg_c)
cosy = bcdotp/(mg_b*mg_c)
if np.ceil(cosx) == np.ceil(cosy):
    print('The value of m is verified')
else:    
    print('The value of m is not true')
    
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)
y_1 = 3*x+1
y_2 = 0.5*x+1.5
y_3 = 1.153*x+4
plt.plot(x,y_1,'-r',label='3x+1')
plt.plot(x,y_2,'-b',label='0.5x+1.5')
plt.plot(x,y_3,'-g',label='1.153x+4')
plt.title('Graph of lines')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
plt.show()    
```
